```python
def extract_names(text):
    try:
        items = ast.literal_eval(text)
        names = [i['name'] for i in items]
        return " ".join(names)
    except:
        return ""
```

```python
movies['genres_clean'] = movies['genres'].apply(extract_names)
movies['keywords_clean'] = movies['keywords'].apply(extract_names)
movies['cast_clean'] = movies['cast'].apply(extract_names)
```

```python
movies['cast_clean'].head()
```

```python
def extract_director(text):
    try:
        items = ast.literal_eval(text)
        names = [i['name'] for i in items if i['job'] == 'Director']
        return " ".join(names)
    except:
        return ""
```

```python
movies['director_clean'] = movies['crew'].apply(extract_director)
```
